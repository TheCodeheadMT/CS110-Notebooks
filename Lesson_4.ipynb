## Strings

Strings are a text data type that is surrounded by either single (```'```) or double (```"```) quotation marks.

### Concatenation

Concatenation is joining strings together and is done by using a plus sign (```+```).  For example:

```
str_1 = "Hello"
str_2 = "World"
str_3 = str_1 + str_2
```

In this case ```str_3``` would end up being ```HelloWorld```.  There is no space between the words because concatenation just combines strings it does not add anything.

### Indexing and Slicing

Indexing is taking a specific character in a string and is done by using square brackets (```[]```) at the end of the string, **the index of all string starts at 0**.  To take the 6th character in ```str_3``` it would look like this ```str_3[5]``` to take the first character it would be ```str_3[0]```.

Slicing is taking a range of characters from the string it works similar to indexing.  Square brackets (```[]```) are still used, but within them the values will be ```[start:stop:step]```.  Start is inclusive and if it is index 0 then nothing is needed, stop is non-inclusive and if the stop is the last character nothing is needed, and step is how you are moving through the string it defaults to 1.
#### For Example:

````python
str_1 = "Hello"
str_2 = "World"
str_3 = str_1 + str_2

print(str_3[0:10])      # Slicing all the characters
print(str_3[:])         # Slicing the whole thing
print(str_3[0:10:2])    # Slicing everyother character
print(str_3[:5])        # Slicing Hello
print(str_3[5:])        # Slicing World
````

### fstrings

fstrings are formatted strings they enable you to insert variables directly into a string using curly brackets (```{}```).  For example if you want to create the sentence "Hi, my name is __, and I am __ years old" where the blanks are your name and age in variables.
With these variables:

```
name = 'Ed'
age = '34'
```

With string it would be: 

```
intro = "Hi, my name is " + name + ", and I am " + age + " years old"
```

With fstring it would be:

```
fintro = "Hi, my name is {name}, and I am {age} years old"
```
**Both of these methods accomplish the same thing it is your choice on what to use.**

## Functions

Functions are a named block of code that performs a specific task. They appear in code as: Function_name(). Within the parentheses, values, known as arguments, are passed to the function for it to perform the desired task on. Depending on the purpose of the function, it might also return a value.

## Print Function

The print function takes the form ```print(outputs, options)```. It takes the outputs and displays them in the console in the labs and inline in these notebooks. ```print()``` does not return any values.

You can have as many outputs as you want. They must be separated by a comma. For example, ```print(Q1, Q2, Q3)``` prints each answer with a space in between.

The space between outputs comes from the separator (```,```) value, which defaults to a space. You can change this using the options. For example, to use a comma and a space: ```print(Q1, Q2, Q3, sep=', ')```. Using *sep=* changes the effect of a comma in the print statement. **This only works in print, not with strings.**

Every print statement ends by default with a newline (```\n```), meaning the next print will be on a different line. This option can also be changed by using ```end=```, just like you would use ```sep=```.

**Keywords to look for in the description, meaning you need to print something, include (This is not all inclusive): Print, Output, Display.**

## Errors (Syntax vs. Semantic)

Syntax errors are coding errors that will stop the program from running.

Examples:

```print("Hello World```

```TF = true```

```ten = 5+'5'```

Semantic errors are coding errors where the code will run but it does not accomplish the intended task.  In this class semantic errors also occure when you don't fully follow instructions, you might get the right answer but not accomplish all that was asked.

Examples:

```
length = 10
width = 5
area = length + width
```

```
value = input()
value = "new value"
print(value)
```

Examples of common Syntax Errors from the console:

| Error | Reason|
|---|---|
| ```NameError: name 'value' is not defined.``` |  Tried to use a variable that has not been assigned a value |
| ```TypeError: can only concatenate str (not "int") to str``` |  Tried to concatenate a string and integer |
| ```TypeError: unsupported operand type(s) for +: 'int' and 'str'``` | Tried to add an integer and string |
| ```TypeError: object of type 'int' has no len()``` | Tried to find the length of a string |
| ```ZeroDivisionError: division by zero``` | Tried to divide by zero |
| ```ValueError: invalid literal for int() with base 10: 'integer'``` | Tried to change the string 'integer' to an integer |
